# Another technique for encoding 

```python
df3.head()
```

```python
columns = ['sex','smoker','region']  # list of column names need to encoded

encoders = {column: LabelEncoder() for column in columns}

for column in columns:
    df3[column] = encoders[column].fit_transform(df3[column])
```
